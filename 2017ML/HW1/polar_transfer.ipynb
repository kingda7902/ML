```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
def cart2polar(x,y):
    x = np.array(x)
    y = np.array(y)
    norm = (x**2+y**2)**0.5
    theta = np.intersect1d(np.arcsin(y/norm),np.arccos(x/norm))
    return (norm, theta)

def polar2cart(r,theta):
    return (r*np.cos(theta),r*np.sin(theta))
```

```python
t = np.pi*np.arange(0.05,0.25,0.025)
data = polar2cart(1,t)
data2 = polar2cart(1,2*t)
data3 = polar2cart(1,(1/t[0])*t**2)
```

```python
plt.tight_layout()
plt.figure(figsize=(5,15))
plt.subplot(311)
plt.plot(data[0], data[1], 'ro')
plt.plot([0,1],[0,1], 'black')
plt.plot([0,1],[0,0], 'black')
plt.subplot(312)
plt.plot(data2[0], data2[1], 'bo')
plt.plot([0,1],[0,1], 'black')
plt.plot([0,1],[0,0], 'black')
plt.subplot(313)
plt.plot(data3[0], data3[1], 'yo')
plt.plot([0,1],[0,1], 'black')
plt.plot([0,1],[0,0], 'black')
```
